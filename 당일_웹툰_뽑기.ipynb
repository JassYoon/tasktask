```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html=urlopen('https://comic.naver.com/webtoon/weekdayList')
soup=BeautifulSoup(html, "lxml")
dailylist=soup.find_all('dt')
del dailylist[0:3]
for wbtn in dailylist:
  print(wbtn.text)
```
